```python
# Noting the time of transaction
import time
# Reading the inventory
fd=open("Inventory.txt","r")

txt=fd.read().split('\n')

fd.close()
# Taking the user input ie the product id

user_name=input("Enter the name : ")

user_phno=input("Enter the phone number : ")

user_mail=input("Enter the emailID : ")

prd_id=input("Enter the product id : ")

# Taking the user input for the number of quantity

qnt_id=input("Enter the quantity : ")

updated_list=[]
# Going through the inventory
for i in txt :
    
    # Going to each of the product in the inventory
    prd_details=i.split(',')
    
    if (prd_details[0]==prd_id):
        
        # If we have sufficient number of products that is to be sold
        
        if (int(qnt_id)<=int(prd_details[3])):
            
            # Updating the inventory
            
            print("-------------------------------------------------")
            
            print("Product ID : ",prd_id)
            
            print("Name of the Product : ",prd_details[1])
            
            print("Price : ",prd_details[2])
            
            print("Quantity Required : ",int(qnt_id))
            
            print("Quantity Present : ",int(prd_details[-1]))
            
            print('=================================================')
            
            print("Total Price : ",int(qnt_id)*int(prd_details[-2]))
            
            print('=================================================')
            
            prd_details[3]=str(int(prd_details[3])-int(qnt_id))
            
            # Generating the sales record
            fd2=open('sales.txt','a')
            sales_details = user_name+","+user_phno+","+user_mail+","+prd_details[1]+","+prd_id+","+qnt_id+","+str(int(qnt_id)*int(prd_details[-2]))+","+time.ctime()+"\n"
            fd2.write(sales_details)
            fd2.close()
        else:
            # If there is not enough product
            
            print("Sorry not enough Materials")
            
            print("We have only ",prd_details[3],'quantity')
            
            print("Would u like to take it ??")
            
            ch=input('Y/N')
            
            if (ch=='Y' or ch=='y'):
                # Updating the inventory and making the number of quantity as 0 as all are sold
                
                print("-------------------------------------------------")
                
                print("Product ID : ",prd_id)
                
                print("Name of the Product : ",prd_details[1])
                
                print("Price : ",prd_details[2])
                
                print("Quantity Required : ",int(prd_details[-1]))
                
                print("Quantity Present : ",int(prd_details[-1]))
                
                print('=================================================')
                
                print("Total Price : ",int(prd_details[-1])*int(prd_details[-2]))
                
                print('=================================================')
                
                fd2=open('sales.txt','a')
                sales_details = user_name+","+user_phno+","+user_mail+","+prd_details[1]+","+prd_id+","+prd_details[-1]+","+str(int(prd_details[-1])*int(prd_details[-2]))+","+time.ctime()+"\n"
                fd2.write(sales_details)
                fd2.close()
                prd_details[3]=str(0)
            else:
                print("Thank You")

    # Appending the details to a new list        
    
    updated_list.append(prd_details)
lst=[]

#Converting the list into a csv format ie comma separated file and appending it to a new list

for i in updated_list:
    prod=i[0]+","+i[1]+","+i[2]+","+i[3]+'\n'
    lst.append(prod)

# Since at every inserting a new line is added we are removing it from the last element

lst[-1]=lst[-1][:-1]

fd1=open('Inventory.txt','w')

# Updating the inventory list by writing it into the file

for i in lst:
    fd1.write(i)
fd1.close()
```
