```python
# Importing time for noting the time the purchase has taken
import time

# Reading the file from the records

fd=open("Records.json","r")

js=fd.read()

fd.close()

# Converting the string data to a json format
record=json.loads(js)

# Creating a menu of the record

print("\t\t\t\t MENU \t\t\t\t")

print("Product ID \tProduct Name \t Product Price \t Product quantity")

for key in record.keys():
    print(str(key)+"\t\t"+record[key]['name']+"\t\t"+str(record[key]['price'])+"\t\t"+str(record[key]['quantity']))
    
# Taking a user data
    
user_name=input("Enter the name         : ")

user_phno=input("Enter the phone number : ")

user_mail=input("Enter the mail id      : ")

prod_id=str(input("Enter the product id : "))

quan_id=int(input("Enter the number of quantity : "))

# There is enough supply then update it

if record[prod_id]['quantity']>=quan_id:

    record[prod_id]['quantity']=record[prod_id]['quantity']-quan_id
    
    print("-----------------------------------------")
    
    print('\t\t BILL')
    
    print("Name of the Product    : "+record[prod_id]['name'])
    
    print("Price of the Product   : "+str(record[prod_id]['price']))
    
    print("Number of quantity     : "+str(quan_id))
    
    print("---------------------------------------------")
    
    print("Total Price            : $"+str(quan_id*record[prod_id]['price']))
    
    if (quan_id*record[prod_id]['price']>2500):
    
        discount =  quan_id*record[prod_id]['price']*0.1
    
        print("Discount               : $"+str(discount))
    
        print("Total Price after discount : $ "+str(quan_id*record[prod_id]['price']-discount))
    
    #Storing the record back to the file
    
    js=json.dumps(record)
    
    fd=open("Records.json","w")
    
    fd.write(js)
    
    fd.close()
    
    #Storing the record of the customer details
    
    fd2=open('SalesReport.txt','a')
    
    if (quan_id*record[prod_id]['price']>2500):
    
        std=str(i)+","+user_name+","+user_phno+","+user_mail+","+record[key]['name']+","+str(record[key]['price'])+","+str(quan_id)+","+str(quan_id*record[prod_id]['price']-discount)+","+time.ctime()+"\n"
    else:
        std=str(i)+","+user_name+","+user_phno+","+user_mail+","+record[key]['name']+","+str(record[key]['price'])+","+str(quan_id)+","+str(quan_id*record[key]['price'])+","+time.ctime()+"\n"
    
    i+=1
    
    fd2.write(std)
    
    fd2.close()
else:
    
    #If there is not enough quantity
    
    print("There is only "+str(record[prod_id]['quantity'])+" products ")
    
    print("Will You take all???")
    
    ch=input("Y/N")
    
    # If all the product is purchased
    
    if (ch=='Y' or ch=='y'):
        
        print("-----------------------------------------")
        
        print('\t\t BILL')
        
        print("Name of the Product    : "+record[prod_id]['name'])
        
        print("Price of the Product   : "+str(record[prod_id]['price']))
        
        print("Number of quantity     : "+str(quan_id))
        
        print("---------------------------------------------")
        
        print("Total Price            : $"+str(record[prod_id]['quantity']*record[prod_id]['price']))
        
        if (record[prod_id]['quantity']*record[prod_id]['price']>2500):
    
            discount =  record[prod_id]['quantity']*record[prod_id]['price']*0.1
    
            print("Discount               : $"+str(discount))
    
            print("Total Price after discount : $ "+str(record[prod_id]['quantity']*record[prod_id]['price']-discount))
        
        #Storing the record of the customer details
        
        fd2=open('SalesReport.txt','a')
        
        if (quan_id*record[prod_id]['price']>2500):
    
            std=str(i)+","+user_name+","+user_phno+","+user_mail+","+record[key]['name']+","+str(record[key]['price'])+","+str(record[prod_id]['quantity'])+","+str(record[prod_id]['quantity']*record[prod_id]['price']-discount)+","+time.ctime()+"\n"
        else:
            std=str(i)+","+user_name+","+user_phno+","+user_mail+","+record[key]['name']+","+str(record[key]['price'])+","+str(record[prod_id]['quantity'])+","+str(record[prod_id]['quantity']*record[prod_id]['price'])+","+time.ctime()+"\n"
        
        i+=1
        
        fd2.write(std)
    
        fd2.close()
        
        # Updating the records 
        
        record[prod_id]['quantity']=0
        
        js=json.dumps(record)
        
        fd=open("Records.json","w")
        
        fd.write(js)
        
        fd.close()
        
    else:
        
        print("Sorry we couldn't be the help. See you Again ")
    
```
