```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
from tqdm import tqdm
```

```python
df=pd.read_csv('quotes.csv')
```

```python
del df['quote']
del df['author_id']
del df['tags']
```

```python
data=[]
for author in tqdm(df.values):
  name=author[0]
  link=author[1]
  # print(author[0]," ",author[1])
  res=requests.get(link)
  soup=BeautifulSoup(res.text,'html.parser')
  dt=soup.find('div',class_='author-details').text
  dob=soup.find('span',class_='author-born-date').text
  loc=soup.find('span',class_='author-born-location').text
  desc=soup.find('div',class_='author-description').text
  data.append(dt)
  fd=open('/content/sample_data/Author'+name+'.txt','w',encoding='utf8')
  fd.write(dt)
  fd.close()

  # print(dt)
  # print('-'*39)
  # print(dob)
  # print('-'*39)
  # print(loc)
  # print('-'*39)
  # print(desc)

```

```python
df['Author_Info']=data
```

```python
data=[]
for i in tqdm(range(1,11)):
  path='https://quotes.toscrape.com/page/'+str(i)+'/'
  req=requests.get(path)
  soup=BeautifulSoup(req.text,'html.parser')
  for sp in soup.find_all('div',class_='quote'):
      tags=[]
      quote = sp.find('span',class_='text').text[1:-1]
      author = sp.find('small').text
      author_id=sp.find('a').get('href')
      tag=sp.find_all('a',class_='tag')
      for i in tag :
          tags.append(i.text)
  #     print(quote,author,author_id,",".join(tags))
      data.append([quote,author,author_id,",".join(tags)])
      # print("Quote : ",quote)
      # print("Author : ",author)
      # print("Author ID : ",author_id)
      # print("Features : ",",".join(tags))
      # print("---"*19)
```

```python
df1=pd.DataFrame(data,columns=['quote','author','author_id','tags'])
```
