```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
dt1=pd.read_csv('googleplaystore.csv')
dt1
```

```python
dt1.shape
```

```python
dt1.isnull().sum()
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(dt1.iloc[:,2:3])
dt1.iloc[:,2:3]=imputer.transform(dt1.iloc[:,2:3])
imputer.fit(dt1.iloc[:,6:7])
dt1.iloc[:,6:7]=imputer.transform(dt1.iloc[:,6:7])
imputer.fit(dt1.iloc[:,8:9])
dt1.iloc[:,8:9]=imputer.transform(dt1.iloc[:,8:9])
imputer.fit(dt1.iloc[:,11:])
dt1.iloc[:,11:]=imputer.transform(dt1.iloc[:,11:])
```
