```python
import requests
import pandas as pd
```

```python
df1=pd.read_csv('stocks.csv')
```

```python
df1.head()
```

##Downloading Image


```python
link=df1['img_link'][0]
res=requests.get(link)
```

```python
print(res.content)
```

```python
file_d=open('img.png','wb')
file_d.write(res.content)
file_d.close()
```

## Downloading all the images

```python
from tqdm import tqdm
```

```python
cnt=1
for i in tqdm(df1['img_link']):
  res=requests.get(i)
  fd=open('/content/sample_data/Images'+'/Images_'+str(cnt)+'.png','wb')
  fd.write(res.content)
  fd.close()
  cnt+=1
```
