## Z test

```python
def ztest(n,sm_mean,pop_mean,std,alternative,alpha=0.05):
  z_score=(sm_mean-pop_mean)/(std/math.sqrt(n))
  p_val=st.norm.sf(z_score)
  print(z_score)
  print(p_val)
  if (alternative=='one-sided'):
    if (p_val<alpha):
      print("Reject the null hypothesis")
    else:
      print("Cannot reject the null hypothesis")
  elif (alternative=='two-sided'):
    if (p_val<alpha/2):
      print("Reject the null hypothesis")
    else:
      print("Cannot reject the null hypothesis")
```

```python
ztest(50,110,100,15,'one-sided')
```

```python
ztest(81,90,82,20,'one-sided')
```

## One-Way Anova

```python
from scipy.stats import f_oneway
```

```python
e=[2,4,2]
m=[2,3,4]
s=[1,2,5]
```

```python
chi_stat,p_val3=f_oneway(e,m,s)
```

```python
if (p_val3<0.05):
  print('Reject Null Hypo')
else:
  print('Fail to reject')
```

```python
performance1=[89,89,88,78,79]
performance2=[93,92,94,89,88]
performance3=[89,88,89,93,90]
performance4=[81,78,81,92,82]
```

```python
chi_stat,p_val4=f_oneway(performance1,performance2,performance3,performance4)
```

```python
if (p_val4<0.05):
  print('Reject Null Hypo')
else:
  print('Fail to reject')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
dataframe = pd.DataFrame({'Fertilizer': np.repeat(['daily', 'weekly'], 15),
                          'Watering': np.repeat(['daily', 'weekly'], 15),
                          'height': [14, 16, 15, 15, 16, 13, 12, 11, 14,
                                     15, 16, 16, 17, 18, 14, 13, 14, 14,
                                     14, 15, 16, 16, 17, 18, 14, 13, 14,
                                     14, 14, 15]})
```

```python
dataframe
```

```python
model=ols('height~ Fertilizer + Watering+Fertilizer:Watering',dataframe).fit()
```

```python
result=sm.stats.anova_lm(model,label=2)
```
