```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a=np.linspace(0,4*np.pi,1000)
sin_val=np.sin(a)
# print(a)
# print('-'*12)
# print(sin_val)
plt.plot(a,sin_val)
plt.title('Sin Wave')
plt.show()
```

```python
b=np.linspace(0,4*np.pi,1000)
cos_val=np.cos(b)
plt.plot(b,cos_val,label='Cos Wave')
plt.plot(a,sin_val,label='Sin Wave')
plt.legend(loc='upper right')
plt.title('Cos Wave and Sin Wave')
plt.show()
```

```python
x=np.linspace(1,50,100)
fig,ax=plt.subplots(2,2)
y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')
ax[0,1].plot(x,np.cos(y))
ax[0,1].set_title('Cos Wave')
ax[1,0].plot(x,np.tan(y))
ax[1,0].set_title('Tan Wave')
ax[1,1].hist(y,bins=100)
ax[1,1].set_title('Random Plot')
plt.tight_layout()
plt.show()
```
