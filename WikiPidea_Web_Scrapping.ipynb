```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
link='https://en.wikipedia.org/wiki/Mahatma_Gandhi'
req=requests.get(link)
soup=BeautifulSoup(req.text,'html.parser')
```

```python
soup
```

```python
heading=soup.find('h1').text
```

```python
heading
```

```python
corpus=''
for p in soup.find_all('p'):
  corpus+=p.text
  # print('--'*20)
  corpus+='\n'
corpus=corpus.strip()
```

```python
print(corpus)
```

```python
for i in range(3,467):
  corpus=corpus.replace('['+str(i)+']','')
```

```python
print(corpus)
```

```python
fd=open('Mahatma_Gandhi.txt','w')
fd.write(corpus)
fd.close()
```
